Import all necessary libraries

```python
import numpy
import matplotlib.pyplot
```

Create a handle for the .txt file containing the histogram generated in the hifiasm log file and filter out for the numbers specifically

```python
original_histogram_file = open("kmer-spectrum.txt")
handle = original_histogram_file.read()

lines = numpy.array(handle.split("\n"))

x_values = numpy.array([])
y_values = numpy.array([])



for line in range(0, len(lines) - 1):
    myLine = lines[line]
    myLine = myLine.replace("*", "")
    myLine = myLine.replace(">","")
    myLine = myLine.replace(" ", "")
    myLine = myLine.replace("[M::ha_hist_line]","")
    myLine = myLine.split(":")
    x_values = numpy.append(x_values, int(myLine[0]))
    y_values = numpy.append(y_values, int(myLine[1]))
```

Now, let's plot the data

```python
matplotlib.pyplot.bar(x_values, y_values, width=1.5, align='edge', color='tab:pink')
matplotlib.pyplot.margins(x=0)
matplotlib.pyplot.xlabel("k-mer")
matplotlib.pyplot.ylabel("frequency")
```
